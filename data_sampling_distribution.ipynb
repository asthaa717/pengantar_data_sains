```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

## Normal Distribution

```python
distribution = {
    "Normal": stats.norm.rvs(loc=0, scale=1, size=100),
}

for name, data in distribution.items():
    x = np.linspace(min(data), max(data), 100)
    pdf = stats.gaussian_kde(data).evaluate(x)  # Estimasi PDF
    cdf = np.cumsum(pdf) / sum(pdf)  # Estimasi CDF
    
    mean = np.mean(data)
    std = np.std(data)
    
    print(f"Distribusi {name}: Mean = {mean :.4f}, Std = {std:.4f}")
    
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Histogram')
    plt.plot(x, pdf, color='red', label='PDF')
    plt.title(f"PDF dari Distribusi {name}")
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(x, cdf, color='green', label='CDF')
    plt.title(f"CDF dari Distribusi {name}")
    plt.legend()
    
    plt.show()
```
